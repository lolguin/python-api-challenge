### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
file = "../output_data/summary.csv"
data = pd.read_csv(file)
data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Access mapgs with unique API key
gmaps.configure(api_key=g_key)
coordinates = zip(data["Lat"], data["Lng"])
coordinates = list(coordinates)

#find max humidity of dataset
max_humidity = data["Humidity"].max()

# Customize the size of the figure
figure_layout = {
    'width': '900px',
    'height': '600px',
    'border': '1px solid white',
    'padding': '1px',
    'margin': '0 auto 0 auto'}

fig = gmaps.figure(layout=figure_layout)

# Assign the marker layer to a variable
markers = gmaps.marker_layer(coordinates)

heat_layer = gmaps.heatmap_layer(coordinates, weights = data["Humidity"], dissipating = False, 
                                 max_intensity = max_humidity, point_radius = 1)
# Add the layer to the map
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
data.head()
```

```python
# Create different filter criterias
temp_filter = (data["Max Temp"] < 80) & (data["Max Temp"] > 70)
wind_filter = data["WindSpeed"] < 10
cloud_filter = data["Cloudiness"] == 0

#create new df based on filters above
my_vacation = data.loc[(temp_filter & wind_filter & cloud_filter), :] 

my_coordinates = zip(my_vacation["Lat"], my_vacation["Lng"])
my_coordinates = list(my_coordinates)

#print filtered dataframe
my_vacation
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# hotel_df = [Range(12)]
my_vacation["Hotel Name"] = ""

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"

params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key, 
    }

for index, row in my_vacation.iterrows():    
    lat = row["Lat"]
    long = row["Lng"]
    params["location"] = f"{lat},{long}"
       
    name_address = requests.get(base_url, params = params)
    name_address = name_address.json()
    my_vacation.loc[index, "Hotel Name"] = name_address["results"][0]["name"]

```

```python
#convert hotel names to list
hotels = my_vacation["Hotel Name"].tolist()
cities = my_vacation["Cities"].tolist()
countries = my_vacation["Country"].tolist()

my_vacation

new_coordinates = zip(my_vacation["Lat"], my_vacation["Lng"])
new_coordinates = list(new_coordinates)
```

```python

# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{Cities}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in my_vacation.iterrows()]
#locations = hotel_df[["Lat", "Lng"]]

hotel_layer = gmaps.marker_layer(
    new_coordinates, 
    info_box_content = hotel_info
)

fig = gmaps.figure()
fig.add_layer(heat_layer)
fig.add_layer(hotel_layer)

fig
```
